```python
# -*- coding: utf-8 -*-
from __future__ import division

import pyomo.environ as pyo, numpy as np, pandas as pd
from pyomo.environ import *
from pyomo.opt import SolverFactory

import matplotlib.pyplot as plt

# Entrada de dados:
#
## Geradores
diretorio = "PIBIC2021"
arquivo = diretorio+"//"+"input.xlsx"
ute = pd.read_excel(arquivo, sheet_name='UTE')
uel = pd.read_excel(arquivo, sheet_name='UEL')
demanda = pd.read_excel(arquivo, sheet_name='demanda')

### termoelétricas:
cos = ute.Custo.values
a = ute.A.values*cos
b = ute.B.values*cos
c = ute.C.values*cos
pmin = ute.Pmin.values
pmax = ute.Pmax.values
ur = ute.UR.values
dr = ute.DR.values

# eólicas:
nt = uel.Nt.values
pt = uel.Pt.values
Wmax = uel.Wmax.values
di = uel.d.values

## Demanda
#Pd = demanda.dem.values
#t = demanda.t.to_list()
Pd = [3000, 3100, 3200, 3300, 3500, 3888, 4300]

Nd = len(Pd)
Ng = len(ute)
Nw = len(uel)

model = pyo.ConcreteModel()

model.Pg = pyo.Var(range(Ng), range(Nd), bounds=(0,None))
Pg = model.Pg
model.Pw = pyo.Var(range(Nw), range(Nd), bounds=(0,None))
Pw = model.Pw

model.deficit = pyo.Var(range(Nd), bounds=(0,None))
deficit = model.deficit

def fg(Pg,a,b,c,t):
    return sum([a[g] + b[g]*Pg[g,t] + c[g]*Pg[g,t]*Pg[g,t] for g in range(Ng)])

'''def gPdw(Wmax):
	L = np.arange(0, Wmax, 0.0001)
	tam = len(L)
	n = random.randint(0,tam)
	np.random.shuffle(L)
	return L[n]'''

def Pdw(v,vi,v0,vr,wr):
    if(v<vi or  v>v0):
        return 0
    elif(vi<=v or v<=vr):
        return wr*((v-vi)/vr-vi)
    else:
        return wr

def Li(Vr,Vi):
    return ((Vr-Vi)/Vi)

def Weibull(Pw,Wmax,Vr,Vi,c,k):
    rho = Pw/Wmax
    l = Li(Vr,Vi)
    return ((k*l*Vi)/c)*((((1+rho*l)*Vi)/c)**(k-1))*exp(-(((1+rho*l)*Vi)/c)**k)

def fw(Pw,di,Wmax,t):
    v = 10
    Vi = 5
    Vr = 15
    V0 = 25
    Kp = 0.95
    Kr = 0.05
    c = 8.47
    k = 2.26

    
    sum_Cl = sum([di[w]*Pw[w,t] for w in range(Nw)])
    sum_Cp = sum([Kp*(Pw[w,t]-Pdw(v,Vi,V0,Vr,Wmax[w]))*Weibull(Pw[w,t],Wmax[w],Vr,Vi,c,k) for w in range(Nw)])
    sum_Cr = sum([Kr*(Pw[w,t]-Pdw(v,Vi,V0,Vr,Wmax[w]))*Weibull(Pw[w,t],Wmax[w],Vr,Vi,c,k) for w in range(Nw)])
    
    return sum_Cl+sum_Cp+sum_Cr

def fun_custo_total(Pg,a,b,c,Pw,di,Wmax,t):
    Fg = fg(Pg,a,b,c,t)
    Fw = fw(Pw,di,Wmax,t)
    return Fg+Fw

# Objetivo
model.obj = pyo.Objective(expr= sum(fun_custo_total(Pg,a,b,c,Pw,di,Wmax,t) + deficit[t]*99999 for t in range(Nd)), sense=minimize)

# Balanco de potencia:
model.balanco = ConstraintList()
for t in range(Nd):
    sum_Pg = sum(Pg[g,t] for g in range(Ng))
    sum_Pw = sum(Pw[w,t] for w in range(Nw))
    model.balanco.add(expr= (sum_Pg + sum_Pw + deficit[t]) == Pd[t])

# limites de geração termo:
model.limger = pyo.ConstraintList()
for g in range(Ng):
    for t in range(Nd):
        model.limger.add(inequality(pmin[g], Pg[g,t], pmax[g]))

# limites de geração eólica:
model.limgerw = pyo.ConstraintList()
for w in range(Nw):
    for t in range(Nd):
        model.limgerw.add(inequality(0, Pw[w,t], Wmax[w]))

# limites de rampas termo
model.ramp_up = pyo.ConstraintList()
for g in range(Ng):
    for t in range(Nd):
        if(t==0):
            Constraint.Skip
        else:
            model.ramp_up.add(expr= Pg[g,t] - Pg[g,t-1] <= ur[g])

model.ramp_down = pyo.ConstraintList()
for g in range(Ng):
    for t in range(Nd):
        if(t==0):
            Constraint.Skip
        else:
            model.ramp_down.add(expr= Pg[g,t-1] - Pg[g,t] <= dr[g])

# limites de rampas eólicas:


opt = SolverFactory('mindtpy')
opt.solve(model, mip_solver='glpk', nlp_solver='ipopt')

model.pprint()

# 
f = open('caso1.csv', 'w')

f.write("t,Pd,")
for g in range(Ng):
    f.write(f"G{g+1}"+",")
for w in range(Nw):
    f.write(f"W{w+1}"+",")
f.write("Gerado,Corte"+"\n")


for t in range(Nd):
    f.write(str(t)+","+str(Pd[t])+",")
    for g in range(Ng):
        f.write(str(pyo.value(Pg[g,t]))+",")
    for w in range(Nw):
        f.write(str(pyo.value(Pw[w,t]))+",")
    f.write(str(sum(pyo.value(Pg[:,t])) + sum(pyo.value(Pw[:,t]))) + "," + str(int(pyo.value(deficit[t]))) + "\n")

f.close()
```
